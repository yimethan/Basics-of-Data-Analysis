# Recursive functions

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)

for i in range(1,10):
    print(factorial(i))
```

## Fib sum functions

```python
%%timeit

import time

def fib_rec(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib_rec(n-1) + fib_rec(n-2)

print(fib_rec(20))
```

### Memoization (top-down approach)

```python
%%timeit

arr = [0]*100

def fib_memo(n):
    # There is a value in the array
    if arr[n-1] != 0:
        return arr[n-1]
    # There is no value in the array
    else:
        if n == 1 or n == 2:
            result = 1
            arr[n-1] = result
            return result
        else:
            result = fib_memo(n-1) + fib_memo(n-2)
            arr[n-1] = result
            return result
        
print(fib_memo(20))    
```

### Tabulation (bottom-up approach)

```python
%%timeit

def fib_tab(n):
    arr = []
    for i in range(n):
        if i == 0 or i == 1:
            arr.append(1)
        else:
            arr.append(arr[i-1] + arr[i-2])
    return arr[n-1]

print(fib_tab(20))
```
